```python
# Concatena verticalmente los dataframes
data_completa = pd.concat([datos_0, datos_1, datos_2], ignore_index=True)
data_completa = data_completa[['Fecha', 'ADC', 'Puerta']]
# Ordena el dataframe por la columna 'Fecha'
data_completa = data_completa.sort_values(by='Fecha')
# Convierte la columna 'Fecha' a formato datetime
data_completa['Fecha'] = pd.to_datetime(data_completa['Fecha'])
# Crea una nueva columna 'Hora_Min_Seg' con la hora, minuto y segundo
data_completa['Fecha'] = data_completa['Fecha'].dt.strftime('%H:%M:%S')
# Selecciona las columnas necesarias
data_completa = data_completa[['Fecha', 'ADC', 'Puerta']]
# Ordena el dataframe por la columna 'Fecha'
data_completa = data_completa.sort_values(by='Fecha')
# Muestra el dataframe resultante
data_completa.head(10000000)


# Convierte la columna 'Fecha' a formato datetime
data_completa['Fecha'] = pd.to_datetime(data_completa['Fecha'], format='%H:%M:%S')
# Filtra los datos para quedarte solo con aquellos con hora mayor a 8:15
data_completa = data_completa[data_completa['Fecha'].dt.time > pd.to_datetime('8:15:00', format='%H:%M:%S').time()]
#De 8:23 a 9:00
vuelta_1 = data_completa[data_completa['Fecha'].dt.time <= pd.to_datetime('9:00:00', format='%H:%M:%S').time()]
vuelta_1['Fecha'] = pd.to_datetime(vuelta_1['Fecha'], format='%H:%M:%S')
#De 9:07 a fin
vuelta_2 = data_completa[data_completa['Fecha'].dt.time >= pd.to_datetime('9:05:00', format='%H:%M:%S').time()]
vuelta_2['Fecha'] = pd.to_datetime(vuelta_2['Fecha'], format='%H:%M:%S')
```

```python
fig = px.scatter(vuelta_1, x='Fecha', y='ADC', color='Puerta', title='Vuelta 1 - Virreyes a Retiro')

# Configuramos el formato de fecha en el eje x
fig.update_xaxes(type='category')

# Mostramos el gráfico interactivo
fig.show()
```
